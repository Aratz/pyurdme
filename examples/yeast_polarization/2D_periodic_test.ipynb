```python
import pyurdme
```

```python
num_vox = 30
class testDiffusion2D(pyurdme.URDMEModel):
    def __init__(self):
        pyurdme.URDMEModel.__init__(self,"test1D")
        X = pyurdme.Species(name="X",  diffusion_constant=0.001)
        self.addSpecies([X])
        self.mesh = pyurdme.Mesh.unitSquareMesh(nx=num_vox, ny=num_vox, periodic=True)
        self.placeNear({X:num_vox*num_vox*10}, (0.1,0.1))
        #self.scatter({X:1000})
        #self.distributeUniformly({X:1000})
        self.timespan(range(101))
```

```python
model = testDiffusion2D()
result = pyurdme.urdme(model)
```

```python
x_vals = model.mesh.coordinates()[:, 0]
y_vals = model.mesh.coordinates()[:, 1]
z_vals = result.getSpecies("X")
```

```python
from matplotlib import cm as CM
from IPython.display import display, clear_output
import time
for tndx in model.tspan:
    fig = plt.figure(figsize=(8,6), dpi=100)
    plt.hexbin(x=x_vals, y=y_vals, C=z_vals[tndx,:], gridsize=num_vox, cmap=CM.jet, bins=None)
    plt.axis([x_vals.min(), x_vals.max(), y_vals.min(), y_vals.max()])
    plt.title("t={0}".format(tndx))
    cb = plt.colorbar()
    cb.set_label('molecule population')
    clear_output()
    plt.show()
    time.sleep(.01)
```
