```python
import pyurdme
```

```python
class testDiffusion1D(pyurdme.URDMEModel):
    def __init__(self):
        pyurdme.URDMEModel.__init__(self,"test1D")
        X = pyurdme.Species(name="X",  diffusion_constant=0.001)
        self.addSpecies([X])
        self.mesh = pyurdme.Mesh.unitIntervalMesh(nx=100, periodic=True)
        self.placeNear({X:1000}, 0)
        #self.scatter({X:1000})
        #self.distributeUniformly({X:1000})
        self.timespan(range(101))
```

```python
model = testDiffusion1D()
result = pyurdme.urdme(model)
```

```python
x_vals = model.mesh.coordinates()[:, 0]
z_vals = result.getSpecies("X")
```

```python
plt.imshow(z_vals)
plt.clim(0,20)
plt.xlabel("space")
plt.ylabel("time")
plt.colorbar()
```
